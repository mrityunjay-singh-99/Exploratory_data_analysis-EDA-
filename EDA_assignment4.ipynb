```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#read the bank data
path=r"C:\Users\Mrityunjay\Desktop\Data science naresh it\Class notes by me\bank.csv"
bank_df=pd.read_csv(path,sep=";")
bank_df
```

```python
colm=bank_df.columns
colm
```

```python
#pick the categorical colomns
cat_colm=[]
for key,value in dict(bank_df.dtypes).items():
    if value=="object":
        cat_colm.append(key)
print(cat_colm)
```

```python
data_frame=r'C:\\Users\\Mrityunjay\\Desktop\\Data science naresh it\\Class Assiments\\value_count_df'
```

```python
#change into dataframe and plot them
for i in cat_colm:
    dfn=bank_df[i].value_counts()
    keys=dfn.keys()
    value=dfn.values
    col=[i,"count"]
    df=pd.DataFrame(zip(keys,value),columns=col)
    df.to_csv(f'{data_frame}\\{i}.csv',index=False)

    plt.bar(i,"count",data=df)
    plt.xlabel(f"{i}")
    plt.ylabel("class frequency")
    plt.savefig(f"{data_frame}\\{i}_bank_df.jpg")
    plt.show()
```
